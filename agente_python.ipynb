```python
from rdflib import Graph, Namespace, URIRef

# =====================================================
# URL RAW DEL ARCHIVO RDF/OWL
# =====================================================

url = "https://raw.githubusercontent.com/alexogis2001/ontologia-recomendador/refs/heads/main/ontologia.rdf"

# =====================================================
# CREAR GRAFO
# =====================================================

g = Graph()

# =====================================================
# CARGAR ONTOLOGÍA
# =====================================================

g.parse(url, format="xml")

# =====================================================
# NAMESPACE
# =====================================================

ns = Namespace("http://www.miOntologia.org/ciberseguridad#")

# =====================================================
# AGREGAR RELACIONES MANUALMENTE
# =====================================================

# FirewallCisco -> mitiga -> Phishing_Corporativo
g.add((
    URIRef(f"{ns}FirewallCisco"),
    URIRef(f"{ns}mitiga"),
    URIRef(f"{ns}Phishing_Corporativo")
))

# Phishing_Corporativo -> afecta -> Servidor_Web
g.add((
    URIRef(f"{ns}Phishing_Corporativo"),
    URIRef(f"{ns}afecta"),
    URIRef(f"{ns}Servidor_Web")
))

# Malware_Troyano -> afecta -> Base_Datos_Empresa
g.add((
    URIRef(f"{ns}Malware_Troyano"),
    URIRef(f"{ns}afecta"),
    URIRef(f"{ns}Base_Datos_Empresa")
))

# Antivirus_Kaspersky -> mitiga -> Malware_Troyano
g.add((
    URIRef(f"{ns}Antivirus_Kaspersky"),
    URIRef(f"{ns}mitiga"),
    URIRef(f"{ns}Malware_Troyano")
))

# =====================================================
# CLASE DEL AGENTE
# =====================================================

class AgenteCiberseguridad:

    def __init__(self, grafo):
        self.grafo = grafo

    # =================================================
    # RECOMENDAR CONTROLES
    # =================================================

    def recomendar(self, amenaza):

        consulta = f"""
        PREFIX ns: <{ns}>

        SELECT ?control
        WHERE {{
            ?control ns:mitiga <{amenaza}> .
        }}
        """

        resultados = self.grafo.query(consulta)

        return [
            str(r[0]).split("#")[-1]
            for r in resultados
        ]

    # =================================================
    # EVALUAR RIESGO
    # =================================================

    def evaluar_riesgo(self, amenaza):

        controles = self.recomendar(amenaza)

        cantidad = len(controles)

        if cantidad == 0:
            return "Riesgo ALTO"

        elif cantidad == 1:
            return "Riesgo MEDIO"

        else:
            return "Riesgo BAJO"

    # =================================================
    # LISTAR AMENAZAS
    # =================================================

    def listar_amenazas(self):

        consulta = f"""
        PREFIX ns: <{ns}>

        SELECT ?amenaza
        WHERE {{
            ?amenaza a ns:Amenaza .
        }}
        """

        resultados = self.grafo.query(consulta)

        return [
            str(r[0]).split("#")[-1]
            for r in resultados
        ]

    # =================================================
    # MOSTRAR ACTIVOS AFECTADOS
    # =================================================

    def mostrar_activos_afectados(self, amenaza):

        consulta = f"""
        PREFIX ns: <{ns}>

        SELECT ?activo
        WHERE {{
            <{amenaza}> ns:afecta ?activo .
        }}
        """

        resultados = self.grafo.query(consulta)

        return [
            str(r[0]).split("#")[-1]
            for r in resultados
        ]


# =====================================================
# CREAR AGENTE
# =====================================================

agente = AgenteCiberseguridad(g)

# =====================================================
# AMENAZA A CONSULTAR
# =====================================================

amenaza = f"{ns}Phishing_Corporativo"

# =====================================================
# MOSTRAR RESULTADOS
# =====================================================

print("===================================")
print("   AGENTE DE CIBERSEGURIDAD")
print("===================================")

# -----------------------------------

print("\nControles recomendados:")

print(
    agente.recomendar(amenaza)
)

# -----------------------------------

print("\nNivel de riesgo:")

print(
    agente.evaluar_riesgo(amenaza)
)

# -----------------------------------

print("\nActivos afectados:")

print(
    agente.mostrar_activos_afectados(amenaza)
)

# -----------------------------------

print("\nAmenazas registradas:")

print(
    agente.listar_amenazas()
)

print("\n===================================")
```
